```python
ca = data.marginalize([0,1])
ns = ca.sample_sizes
pts = 60

fs = single_pop(ns, pts)

pylab.figure(figsize=(9, 9), dpi=300)
dadi.Plotting.plot_1d_comp_multinom(ca, fs)
pylab.show()
```

```python
def single_pop(ns, pts):
    xx = dadi.Numerics.default_grid(pts)
    phi = dadi.PhiManip.phi_1D_snm(xx)
    fs = dadi.Spectrum.from_phi(phi, ns, (xx,))
    return fs
```
