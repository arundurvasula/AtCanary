# Q1) Where did the Canary Island Arabidopsis split from?

```python
%matplotlib inline
import dadi
import numpy
import pylab
import demographic_models
import migration_models

from numpy import array
```

```python
data = dadi.Spectrum.from_file('../data/matrix_dadi_canarianSubset.transp.genodadiSpectrum.txt')
data = data.fold()
```

```python
pylab.figure()
dadi.Plotting.plot_3d_spectrum(data,vmin=1,pop_ids=("Ca","Mo","Ib"))
pylab.show()
```

# General migration model
## Pop 2 splits into 2 and 3

    nuIb: Iberian population size
    nuMo: Morrocan population size
    nuCa: Canary island population size
    T1: split time for first event
    T2: split time for second event
    m12: migration rate from pop 1 to pop 2
    m21: migration rate from pop 2 to pop 1
    m13: migration rate from pop 1 to pop 3
    m23: migration rate from pop 2 to pop 3

```python
ns = data.sample_sizes
pts_l = [40, 50, 60]
params = array([2.5, 1, 0.8, 0.2, 0.7, 0.1, 0.2, 0.1, 0.1])
upper_bound = [1e1, 1e2, 1e1, 1e1, 1e1, 1e1, 1e1, 1e1, 1e1]
lower_bound = [1e-3, 1e-1, 1e-2, 1e-4, 1e-4, 1e-3, 1e-3, 1e-3]

func = migration_models.split_2_3_migration
func_ex = dadi.Numerics.make_extrap_log_func(func)
```

```python
gm_model = func_ex(params, ns, pts_l)
ll_gm_model = dadi.Inference.ll_multinom(gm_model, data)
print 'Model log-likelihood:', ll_gm_model
theta = dadi.Inference.optimal_sfs_scaling(gm_model, data)
```

```python
p0 = dadi.Misc.perturb_params(params, fold=1, upper_bound=upper_bound)
popt = dadi.Inference.optimize_log(p0, data, func_ex, pts_l,
                                   lower_bound=lower_bound,
                                   upper_bound=upper_bound,
                                   verbose=len(params),
                                   maxiter=3)
print 'Optimized parameters', repr(popt)
gm_model = func_ex(popt, ns, pts_l)
ll_opt = dadi.Inference.ll_multinom(gm_model, data)
print 'Optimized log-likelihood:', ll_opt
```

```python
#nuIb, nuMo, nuCa, T1, T2, m12, m21, m13, m23
pylab.figure(figsize=(9, 9), dpi=300)
dadi.Plotting.plot_3d_comp_multinom(gm_model, data, vmin=1, resid_range=3,
                                    pop_ids =('Ca','Mo', 'Ib'))
pylab.show()
```

```python
def detailed((nuIb, nuMo, nuCaB, nuCaF, TIb, TMo, TCa, mIbMo, mMoIb, mIbCa, mMoCa),
    ns, pts):

    xx = dadi.Numerics.default_grid(pts)
    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.Integration.one_pop(phi, xx, TIb, nu=nuIb)
    
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi)
    phi = dadi.Integration.two_pops(phi, xx, TMo, nu1=nuIb, nu2=nuMo, m12=mIbMo, m21=mMoIb)
    
    phi = dadi.PhiManip.phi_2D_to_3D_split_2(xx, phi)
    
    nuCa_func = lambda t: nuCaB*(nuCaF/nuCaB)**(t/TCa)
    
    phi = dadi.Integration.three_pops(phi, xx, TCa, nu1=nuIb, nu2=nuMo, nu3=nuCa_func, m12=mIbMo, m21=mMoIb, m13=mIbCa, m23=mMoCa)
    
    fs = dadi.Spectrum.from_phi(phi,  ns, (xx,xx,xx))
    
    return fs
```

```python
ns = data.sample_sizes
pts_l = [40, 50, 60]
# nuIb, nuMo, nuCaB, nuCaF, TIb, TMo, TCa, mIbMo, mMoIb, mIbCa, mMoCa
params = array([2.5, 1, 0.2, 1, 1, 1, 1, 0.5, 0.5, 1, 0.5]) #change these?
upper_bound = [100, 100, 10, 100, 10, 10, 10, 50, 50, 50, 50] #change these?
lower_bound = [1e-2, 1e-2, 1e-4, 1e-2, 1e-3, 1e-3, 1e-3, 0, 0, 0, 0] #change these?

func = detailed
func_ex = dadi.Numerics.make_extrap_log_func(func)
```

```python
model = func_ex(params, ns, pts_l)
ll_model = dadi.Inference.ll_multinom(model, data)
print 'Model log-likelihood:', ll_model
theta = dadi.Inference.optimal_sfs_scaling(model, data)
```

```python
p0 = dadi.Misc.perturb_params(params, fold=1, upper_bound=upper_bound)
popt = dadi.Inference.optimize_log(p0, data, func_ex, pts_l,
                                   lower_bound=lower_bound,
                                   upper_bound=upper_bound,
                                   verbose=len(params),
                                   maxiter=1)
print 'Optimized parameters', repr(popt)
model = func_ex(popt, ns, pts_l)
ll_opt = dadi.Inference.ll_multinom(model, data)
print 'Optimized log-likelihood:', ll_opt
```

```python
pylab.figure(figsize=(9, 9), dpi=300)
dadi.Plotting.plot_3d_comp_multinom(model, data, vmin=1, resid_range=4.5,
                                    pop_ids =('Ca','Mo', 'Ib'))
pylab.show()
```
